```python
import numpy as np;
import matplotlib.pyplot as plt;
import pandas as pd;
```

```python
dataset = pd.read_csv("./INvideos.csv")
print(dataset);

print(dataset.columns.values);
```

```python

print("Columns before dropping:", dataset.columns.values)

columns_to_drop = ['video_id', 'thumbnail_link']
dataset = dataset.drop(columns=[col for col in columns_to_drop if col in dataset.columns])

print("Columns after dropping:", dataset.columns.values)
print(dataset[:3])
```

```python
## remove all the rows with video_error_or_removed = True
dataset = dataset[dataset.video_error_or_removed == False]
dataset = dataset.drop(columns=['video_error_or_removed'])
print(dataset[:3])
```

```python
# Handling Missing Values
print(dataset.isnull().sum())
dataset = dataset.dropna()
print(dataset.isnull().sum())
```

```python
## Convert trending_date and publish_time to datetime
dataset['trending_date'] = pd.to_datetime(dataset['trending_date'], format='%y.%d.%m')
dataset['publish_time'] = pd.to_datetime(dataset['publish_time'], format='%Y-%m-%dT%H:%M:%S.%fZ')
print(dataset[:3])
```

```python
## split X,Y : Independent and Dependent Variables
X = dataset.drop(columns=['views'])
Y = dataset['views']

print(X[:3])
print(Y[:3])
```

```python
# Encoding Categorical Data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder = LabelEncoder()
X['channel_title'] = labelencoder.fit_transform(X['channel_title'])
X['category_id'] = labelencoder.fit_transform(X['category_id'])
print(X[:3])
```

```python
print(Y[:3]);
```

```python
# Split the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
print(X_train[:3])
print(X_test[:3])
print(Y_train[:3])
print(Y_test[:3])
```

```python
import numpy as np
```

```python
features=np.array([
    [0,0]
    ,[0,1]
    ,[1,0]
    ,[1,1]
])
labels= np.array([0, 0, 0, 1])
print('Features:', features)
print('Labels:', labels)
```

```python
# epoch(Training Iterations)
# Bias=1, Learning Rate=(0to1)
# input Weights (w1,w2)
epochs = int(input('Enter the Epochs:'))
threshold= float(input('Enter the Threshold:'))
Learning_rate = float(input('Enter the Learning Rate'))
a=[]
a.append(float(input('Enter the weights for x0')))
a.append(float(input('Enter the weights for x1')))
w=np.array(a)
print('Initial Weights:', w)
```

```python
for epoch in range(epochs):
    print(f'Epoch {epoch+1}/{epochs}')
    for i in range(len(features)):
        x = features[i]
        y = labels[i]
        y_pred = np.dot(x, w) >= threshold
        error = y - y_pred
        print(f'  Sample {i+1}: x={x}, y={y}, y_pred={y_pred}, error={error}')
        w += Learning_rate * error * x
        print(f'  Updated Weights: {w}')
```

```python
print('Final Weights:', w)
```
